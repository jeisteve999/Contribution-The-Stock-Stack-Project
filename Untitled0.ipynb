```python
import pandas as pd

df = pd.read_csv('Amazon_sales_Jversion.csv', sep=';', encoding='utf-8', on_bad_lines='warn')
```

```python
df.info()
df.head()
```

```python
# Verificar valores nulos por columna
df.isnull().sum()

# Eliminar filas con valores nulos
df = df.dropna()

# Eliminar columnas que no aportan
df = df.drop(columns=['Unnamed: 22'])
```

```python
# Print de la verificacion
print("Dimensiones tras limpieza:", df.shape)
```

```python
# Análisis descriptivo de Amount y Qty
import numpy as np
from scipy.stats import skew, kurtosis

for col in ['Amount', 'Qty']:
    print(f"\n📊 Estadísticas para {col}:\n")
    serie = df[col]

    print("Media:", serie.mean())
    print("Mediana:", serie.median())
    print("Moda:", serie.mode().values)
    print("Mínimo:", serie.min())
    print("Máximo:", serie.max())
    print("Rango:", serie.max() - serie.min())
    print("Desviación estándar:", serie.std())
    print("Varianza:", serie.var())
    print("Asimetría (skew):", skew(serie))
    print("Curtosis:", kurtosis(serie))

    Q1 = serie.quantile(0.25)
    Q2 = serie.quantile(0.50)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1
    print("Q1:", Q1)
    print("Q2 (mediana):", Q2)
    print("Q3:", Q3)
    print("IQR:", IQR)

    # Outliers
    outliers = df[(serie < Q1 - 1.5 * IQR) | (serie > Q3 + 1.5 * IQR)]
    print("Cantidad de outliers:", len(outliers))
```

```python
# Graficando distribuciones
import seaborn as sns
import matplotlib.pyplot as plt

for col in ['Amount', 'Qty']:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir columna 'Date' a tipo fecha
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Verificar fechas mal parseadas
print("Fechas mal parseadas:", df['Date'].isnull().sum())

# Crear columna con el mes
df['Mes'] = df['Date'].dt.to_period('M')

# Filtrar datos desde enero 2022 hasta enero 2023
inicio = pd.Period('2022-01', freq='M')
fin = pd.Period('2023-01', freq='M')
df_filtrado = df[(df['Mes'] >= inicio) & (df['Mes'] <= fin)]

# Agrupar por mes y sumar
ventas_mensuales = df_filtrado.groupby('Mes')['Amount'].sum()

# Convertir índices de Periodo a string para graficar
meses = ventas_mensuales.index.astype(str)
valores = ventas_mensuales.values

# Graficar
plt.figure(figsize=(12, 5))
barras = plt.bar(meses, valores, color='skyblue', edgecolor='black')
plt.title('Ventas totales por mes (2022 - Ene 2023)')
plt.xlabel('Mes')
plt.ylabel('Monto total vendido')
plt.xticks(rotation=45)

# Agregar etiquetas sobre cada barra
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura,
             f'{int(altura):,}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar por categoría y día de la semana
categoria_dia = df.groupby(['Category', 'DíaSemana'])['Amount'].sum().reset_index()

# Ordenar los días
orden_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
categoria_dia['DíaSemana'] = pd.Categorical(categoria_dia['DíaSemana'], categories=orden_dias, ordered=True)

plt.figure(figsize=(12, 6))
sns.lineplot(data=categoria_dia, x='DíaSemana', y='Amount', hue='Category', marker='o')
plt.title('Ventas por categoría y día de la semana')
plt.ylabel('Monto vendido')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
categoria_mes = df.groupby(['Category', 'Mes'])['Amount'].sum().reset_index()

# Orden de meses
orden_meses = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
categoria_mes['Mes'] = pd.Categorical(categoria_mes['Mes'], categories=orden_meses, ordered=True)

plt.figure(figsize=(12, 6))
sns.lineplot(data=categoria_mes, x='Mes', y='Amount', hue='Category', marker='o')
plt.title('Ventas por categoría y mes')
plt.ylabel('Monto vendido')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
